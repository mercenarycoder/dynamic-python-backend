```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing the dataset

```python
dataset = pd.read_csv("./retail_sales.csv")
#Manually creating a array to track which are the Numerical columns from the given DATA
numeric_columns = ['Quantity','Price','Sales']
#this will convert the columns to numeric data type which removes
#error due to absence of data
for col in numeric_columns:
    dataset[col] = pd.to_numeric(dataset[col], errors='coerce')
    
print(dataset.dtypes)

numerical_cols = dataset.select_dtypes(
    include=['int64','float64']
).columns 

categorical_cols = dataset.select_dtypes(
    include=['object','str']
).columns

X = dataset.iloc[:,1:5].values
Y = dataset.iloc[:,[0,5,6,7]].values
Z = dataset.iloc[:, -1].values

print(Y)
print(Z)
```

Fill the missing data so that it can be consistent

```python
from sklearn.impute import SimpleImputer

data = np.array(Y)
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
Y = imputer.fit_transform(data)
np.set_printoptions(suppress=True)
print(X)
print(Y)
```

Encoding the Independent Variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
 
ct = ColumnTransformer(transformers=[('encoder', 
                       OneHotEncoder(),
                       [0,1,2,3])],
                       remainder='passthrough')
X = ct.fit_transform(X).toarray()
print(X)
```

Encoding the Dependent Variable

```python
from sklearn.preprocessing import  LabelEncoder

le = LabelEncoder()
Z = le.fit_transform(Z)
print(Z) 
```

Splitting the dataset into the Training and Test set

```python
from sklearn.model_selection import train_test_split

x_final = np.concatenate((X,Y),axis=1) #axis 1 will combine them horizontally
#print(x_final)

X_train, X_test, Y_train, Y_test = train_test_split(
    x_final, #independent data
    Z,       #dependent data yes,no types
    test_size=0.2,
    random_state= 1
)
```
